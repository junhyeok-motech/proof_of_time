```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("shanchen/Accepted-Papers-Aggregated")
```

```python
ds
```

```python
# keep only 'title', 'authors', 'abstract' and 'tags' columns
ds = ds.select_columns(['title', 'authors', 'abstract', 'tags'])
ds
# keep only the rows where 'tags' is not None
ds = ds.filter(lambda x: x['tags'] is not None)
ds

```

```python
# Check what splits are available
print("Available splits:", ds.keys())

# Option 1: Convert a specific split
# df = ds['train'].to_pandas()

# Option 2: Combine all splits into one DataFrame
import pandas as pd
df = pd.concat([split.to_pandas() for split in ds.values()], ignore_index=True)
df
```

```python
df['tags'] = [i.split(',')[0] for i in df['tags']]
df
```

```python
df.to_csv('accepted_papers.csv', index=False)
```
